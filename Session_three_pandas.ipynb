# Merge, join and concat

```python
df1 = pd.DataFrame({
    'name': ['ali', 'mohammad', 'hassan'],
    'job': ['Programmer', 'teacher', 'taxi driver'],
    'age': [23, 24, 25]
})

df2 = pd.DataFrame({
    'name': ['ali', 'mohammad', 'hassan', 'arian'],
    'job': ['Programmer', 'teacher', 'taxi driver', 'police officer'],
    'weight': [60, 70, 80, 90]
})
```

```python
df1
```

```python
df2
```

### Merge()

#### Inner

```python
pd.merge(df1, df2)
```

```python
pd.merge(df1, df2, on=['job'])
```


#### Outer

```python
pd.merge(df1, df2, how='outer')
```

#### Left outer

```python
pd.merge(df1, df2, how='left')
```

#### Right outer

```python
pd.merge(df1, df2, how='right')
```

## Example 1

```python
df1 = pd.DataFrame({'key': ['b', 'b', 'a', 'c', 'a', 'a', 'b'], 'data1': range(7)})
df2 = pd.DataFrame({'key': ['a', 'b', 'd'], 'data2': range(3)})
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1, df2)
```

```python
pd.merge(df1, df2, how='outer')
```

```python
pd.merge(df1, df2, how='left')
```

```python
pd.merge(df1, df2, how='right')
```

## Example 2

```python
df3 = pd.DataFrame({'lkey': ['b', 'b', 'a', 'c', 'a', 'a', 'b'], 'data1': range(7)})
df4 = pd.DataFrame({'rkey': ['a', 'b', 'd'], 'data2': range(3)})
```

```python
df3
```

```python
df4
```

```python
pd.merge(df3, df4, left_on='lkey', right_on='rkey')
```

## Merging on Index

```python
left1 = pd.DataFrame({'key': ['a', 'b', 'a', 'a', 'b', 'c'], 'value': range(6)})
right1 = pd.DataFrame({'group_val': [3.5, 7]}, index=['a', 'b'])
```

```python
left1
```

```python
right1
```

```python
pd.merge(left1, right1, left_on='key', right_index=True)
```

### Join()

```python
left2 = pd.DataFrame([[1., 2.], [3., 4.], [5., 6.]], index=['a', 'c', 'e'], columns=['Ohio', 'Nevada'])
right2 = pd.DataFrame([[7., 8.], [9., 10.], [11., 12.], [13, 14]], index=['b', 'c', 'd', 'e'], columns=['Missouri', 'Alabama'])
```

```python
left2
```

```python
right2
```

```python
pd.merge(left2, right2, how='inner', left_index=True, right_index=True)
```

```python
pd.merge(left2, right2, how='outer', left_index=True, right_index=True)
```

```python
left2.join(right2, how='outer')
```

### Concat()

```python
import numpy as np
```
